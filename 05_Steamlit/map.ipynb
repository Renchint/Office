```python
import pandas as pd
import folium
```

```python
df_val = pd.read_excel('D:/Projects/Office/04_Excel/Negtgel_office.xlsx', sheet_name='2', header=1)
df_val.info()
```

```python
import folium

# Газрын зургийн төв координат
map_center = [47.918873, 106.917701]
m = folium.Map(location=map_center, zoom_start=13)

# Marker нэмэх
for idx, row in df_val.iterrows():
    coord = row['Координат']
    if isinstance(coord, str) and ',' in coord:
        try:
            lat, lon = map(float, coord.split(','))
            
            # Popup мэдээлэл
            popup_text = f"""
            <b>Оффис:</b> {row.get('Барилгын нэр', '')}<br>
            <b>Дүүрэг:</b> {row.get('Дүүрэг', '')}<br>
            <b>Хороо:</b> {row.get('Хороо', '')}<br>
            <b>Байрны дугаар:</b> {row.get('Байрны дугаар', '')}<br>
            <b>Ашиглалтад орсон:</b> {row.get('Ашиглалтад орсон он', '')}<br>            
            <b>Давхар:</b> {row.get('Нийт давхарын тоо', '')}<br>
            <b>Үнэлгээ:</b> {row.get('Үнэлгээ', '')}
            """
        

            folium.Marker(
                location=[lat, lon],
                popup=folium.Popup(popup_text, max_width=300),
                icon=folium.Icon(color='blue', icon='building')
            ).add_to(m)
        except ValueError:
            print(f"Invalid coordinate: {coord}")

# Газрын зургийг HTML файлд хадгалах
m.save("Valuation_live.html")
```
